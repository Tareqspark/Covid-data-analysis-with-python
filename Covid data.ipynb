```python
df = df[~(df.Confirmed < 10)] 
# To remove the records satisfying a particular condition
```

In which Region, maximum number of Confirmed cases were recorded ?

```python
df.groupby("Region").Confirmed.sum().sort_values(ascending=False)
#First we have calculated the total confirmed values and shorted them , thus we got the hight pick
```

In which Region, minimum number of Deaths cases were recorded ?

```python
df.groupby("Region").Confirmed.sum().sort_values(ascending= True)
```
